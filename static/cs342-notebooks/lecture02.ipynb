## Indexing, Slicing, and Broadcasting

Tensors can be indexed in a standard way, by providing one index for each dimension, but we can also use "wildcard" indices to take access whole sections of the tensor at once.

```python
a = torch.arange(10)
```

```python
a[4]
```

```python
a.shape
```

```python
a[None].shape
```

```python
a[:,None].shape
```

```python
b = torch.zeros((4, 5))
```

```python
b.shape
```

```python
b[2, 3]
```

```python
b[:, 3]
```

```python
b[2, :]
```

```python
b[:,None].shape
```

```python
b[:,:,None].shape
```

Broadcasting allows us to apply operations on tensors of different dimensions, as long as their trailing dimensions match or are one.

```python
b = torch.arange(10) * 10
print(a, b)
```

```python
a + b
```

```python
b + 5
```

```python
torch.ones((1, 3)) + torch.ones((3, 1))
```

```python
torch.ones((2, 3)) + torch.ones((3, 2))
```

```python
a[None] + b[:,None]
```

```python
a[None] * b[:,None]
```
